```python
from WanPy.pythTB_wan import Wannier, Bloch, K_mesh
import WanPy.models as models
import numpy as np
```

```python
delta = 1
t0 = 0.4
tprime = 0.5

model = models.chessboard(t0, tprime, delta).make_supercell([[2,0], [0,2]])
low_E_sites = np.arange(0, model.get_num_orbitals(), 2)
```

```python
random = False

omit_sites = 0
tf_list = list(np.setdiff1d(low_E_sites, [omit_sites])) # delta on lower energy sites omitting the last site

if random:
    n_tfs = 3
    tf_list = "random"

omit_sites = 0
tf_list = list(np.setdiff1d(low_E_sites, [omit_sites])) # delta on lower energy sites omitting the last site
WFs = Wannier(model, tf_list, [20, 20])
```

```python
WFs.plot(0)
```

```python
n_orb = model.get_num_orbitals()
n_occ = int(n_orb/2)
WFs.max_loc(outer_state_idxs=list(range(n_occ)), verbose=True, iter_num_omega_i=100, iter_num_omega_til=10000)
```
