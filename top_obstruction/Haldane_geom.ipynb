```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch

# Define lattice vectors
a1 = np.array([1, 0])  # First lattice vector
a2 = np.array([0.5, np.sqrt(3) / 2])  # Second lattice vector

# Define nearest-neighbor vectors
nn_vectors = [a1, a2, -a1 + a2]

# Define next-nearest-neighbor vectors
nnn_vectors = [a1 + a2, -a1, -a2]

# Create the figure
fig, ax = plt.subplots(figsize=(6, 6))

# Define the positions of the A and B sublattice atoms within the unit cell
A_sublattice = np.array([0, 0])  # A sublattice position
B_sublattice = np.array([0.5, np.sqrt(3) / 6])  # B sublattice position

# Plot the sublattice points
ax.plot(A_sublattice[0], A_sublattice[1], 'ro', label="A Sublattice", markersize=10)
ax.plot(B_sublattice[0], B_sublattice[1], 'bo', label="B Sublattice", markersize=10)

# Function to plot arrows
def plot_arrow(ax, start, direction, color, label=None):
    arrow = FancyArrowPatch(posA=start, posB=start + direction, 
                            arrowstyle='->', mutation_scale=15, color=color)
    ax.add_patch(arrow)
    if label:
        # Add label at the midpoint of the arrow
        mid_point = start + 0.5 * direction
        ax.text(mid_point[0], mid_point[1], label, fontsize=12, color=color)

# Plot the nearest-neighbor hopping with arrows (red and blue)
for i, nn in enumerate(nn_vectors):
    color = 'r' if i % 2 == 0 else 'b'
    plot_arrow(ax, A_sublattice, nn, color)
    plot_arrow(ax, B_sublattice, -nn, color)  # Reverse direction for the B sublattice

# Plot the next-nearest-neighbor hopping with dashed lines and complex hopping
for nnn in nnn_vectors:
    plot_arrow(ax, A_sublattice, nnn, 'k', label=r'$it_2$')

# Add lattice labels for clarity (A, B, and vectors b1, b2, b3)
ax.text(A_sublattice[0] + 0.1, A_sublattice[1] + 0.1, "A", fontsize=16, color="red", fontweight='bold')
ax.text(B_sublattice[0] + 0.1, B_sublattice[1] + 0.1, "B", fontsize=16, color="blue", fontweight='bold')
ax.text(A_sublattice[0] + 1.1, A_sublattice[1], r"$b_1$", fontsize=14, color="black")
ax.text(A_sublattice[0] - 0.9, A_sublattice[1] + 0.9, r"$b_2$", fontsize=14, color="black")
ax.text(A_sublattice[0] + 0.6, A_sublattice[1] - 1.1, r"$b_3$", fontsize=14, color="black")

# Set axis limits and labels
ax.set_xlim([-1.5, 2])
ax.set_ylim([-1.5, 2])
ax.set_aspect('equal')

# Add legend and grid
ax.legend()
plt.grid(True)
plt.title('Haldane Model: Nearest and Next-Nearest Neighbor Hopping')

# Show the plot
plt.show()
```
