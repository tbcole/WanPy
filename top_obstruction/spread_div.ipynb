```python
import pickle

import os
import sys
print(os.getcwd())
sys.path.append("../WanPy")

import pythTB_wan
import matplotlib.pyplot as plt
import numpy as np
```

```python
nkx, nky = 30, 30
n_wfs = 4

# file_name = f"data/WFs_{nkx}x{nky}_t0_list.pkl"
file_name = f"/Users/treycole/Codes/WanPy/dev notebooks/data/n_tfs_{n_wfs}_2_prim_cells_{nkx}x{nky}_k_mesh_fxn_of_t0_WFs.npy"
WFs = np.load(file_name, allow_pickle=True).item()
n_wfs = len(WFs[0.1].spread)
t0_list = list(WFs.keys())
omega_i_list = [val.omega_i/n_wfs for _, val in WFs.items()]
omega_til_list = [val.omega_til/n_wfs for _, val in WFs.items()]

print(f"# Wannier fxns: {n_wfs}")
```

```python
fig, ax = plt.subplots()

plt.scatter(t0_list, omega_i_list)
ax.scatter(t0_list, omega_i_list, label=fr'$\Omega_I / {n_wfs}$', c='r')
ax.scatter(t0_list, omega_til_list, label=fr'$\tilde{{\Omega}} / {n_wfs}$', c='b')
ax.set_xlabel(fr'$t_0$')
ax.set_title(f"{n_wfs}/4 Wannier functions. {nkx}x{nky} k-mesh")
ax.legend()

# plt.savefig("../Images/prim_cell.png")
plt.savefig("../Images/spread_vs_t0_doubled_cell_30x30_k_mesh.png")
```
