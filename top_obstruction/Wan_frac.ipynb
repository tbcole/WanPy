```python
import os
import sys
print(os.getcwd())
sys.path.append("../WanPy")

import pythTB_wan
import matplotlib.pyplot as plt
import numpy as np
```

```python
delta=1
t=1
t2=-0.4
n_occ=25
WFs = np.load(
   f"/Users/treycole/Codes/WanPy/top_obstruction/data/Wan_frac_C=1.0_Delta={delta}_t={t}_t2={t2}_n_occ={n_occ}_WFs.npy",
    allow_pickle=True
    )
WFs_dict = WFs.item()
WFs_dict
```

```python
n_WFs = np.array(list(WFs_dict.keys()))
n_occ = n_WFs[0]+1
WF_fill_frac = n_WFs/n_occ
omega_til_lst = np.array([wf.omega_til for wf in list(WFs_dict.values())])
omega_i_lst = np.array([wf.omega_i for wf in list(WFs_dict.values())])
avg_omega_til = omega_til_lst / n_WFs
avg_omega_i = omega_i_lst / n_WFs

print(f"# occupied bands: {n_occ}")
```

```python
plt.scatter(WF_fill_frac, avg_omega_til, c='r', s=15, label=r'$\langle \widetilde{\Omega} \rangle$') # / # Wannier functions
plt.scatter(WF_fill_frac, avg_omega_i, c='b', s=15, label=r'$\langle \Omega_I \rangle$')

# plt.ylabel(r'$\langle \widetilde{\Omega} \rangle$', size=12)
plt.xlabel(r'$n_{Wf}\ /\ n_{\rm occ}$')
# plt.title(r"C=-1.0 $\Delta$ = 1 $t_0=0.4$ $t'=0.5$")
plt.ylim(-0.01, .28)
plt.xlim(0, 1.0)
plt.legend()
save_name = f'Delta={delta}_t={t}_n_occ={n_occ}'
plt.savefig(f'images/Wan_frac_{save_name}.png', dpi=700)
```
